```python
import bs4, requests
import csv
from google.colab import files
import chardet

def get_product_price(product_url):
    try:
        response = requests.get(product_url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {product_url}: {e}")
        return None

    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    price_elements = soup.select('body > main > div:nth-child(8) > div.container > div > div.col-12.col-lg-9 > div:nth-child(1) > div.col-12.col-xl-5.mb-4 > div > div:nth-child(3) > div:nth-child(1) > strong')
    
    if not price_elements:
        return "Price element not found"

    return price_elements[0].text.strip()

def read_urls_from_csv(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    file_encoding = result['encoding']

    with open(file_path, mode='r', encoding=file_encoding) as csv_file:
        csv_reader = csv.reader(csv_file)
        urls = [row[0] for row in csv_reader]
    return urls

uploaded = files.upload()
csv_file_path = list(uploaded.keys())[0]

urls = read_urls_from_csv(csv_file_path)

for url in urls:
    price = get_product_price(url)
    if price is not None:
        print(f'The price at {url} is {price}')
```
